```python
import nltk
import urllib
```

```python
with urllib.request.urlopen('http://python.org/') as response:
    html = response.read()
```

```python
html[:100]
```

```python
from bs4 import BeautifulSoup
```

```python
# to strip away the HTML code 
web_str = BeautifulSoup(html).get_text()
```

```python
web_str[:100]
```

```python
print(web_str)
```

```python
start = web_str.find("Applications")
```

```python
end = web_str.find("All releases")
```

```python
last_sent = len("All releases")
```

```python
intro = web_str[start:end+last_sent]
```

```python
print(intro)
```

```python
intro_tokens = nltk.word_tokenize(intro)
```
