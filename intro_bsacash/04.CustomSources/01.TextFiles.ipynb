```python
import nltk
```

```python
narek = open("../03.Tokenization/narek.txt", "r", encoding="utf-8").read()
```

```python
narek
```

```python
print(narek)
```

```python
narek_tokens = nltk.word_tokenize(narek)
```

```python
narek_ws = [word for word in narek_tokens if word.isalpha()]
```

```python
narek_fd = nltk.FreqDist(narek_ws)
```

```python
narek_fd.most_common(10)
```
