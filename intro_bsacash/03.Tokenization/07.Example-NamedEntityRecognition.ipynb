```python
import nltk
```

```python
text = open("example.txt").read()
```

```python
text_tag = nltk.pos_tag(nltk.word_tokenize(text))
```

```python
text_ch = nltk.ne_chunk(text_tag)
```

```python

print(text_ch)
```

```python
for  chunk in text_ch:
    if hasattr(chunk, 'label'):
        print(chunk.label(), " ".join(c[0] for c in chunk.leaves()))
```

```python
for  chunk in text_ch:
    if hasattr(chunk, 'label'):
        print(chunk.leaves())
```
